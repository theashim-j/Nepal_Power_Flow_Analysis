```python
import math

# assume a 0.9 lagging power factor for every load
pf = 0.9

def add_load(bus_key, p_mw, name):
    """
    Create a load at the bus identified by bus_key in the `buses` dict,
    with active power p_mw and reactive power computed from pf.
    """
    # if you really want a true zero‐MW load, you can pass p_mw=0
    p_used = max(p_mw, 1e-3)
    q_mvar = p_used * math.tan(math.acos(pf))
    pp.create_load(
        net,
        bus    = buses[bus_key],
        p_mw   = p_mw,
        q_mvar = round(q_mvar, 3),
        name   = name
    )


add_load("attariya",       350, name="Attariya Load")
#add_load("lumki",          1, name="Lumki Load")
add_load("kohalpur",         200, name="Kohalpur Load")
add_load("lamahi",           250, name="Lamahi Load")
add_load("butwal_220",         1660, name="Butwal Load")
add_load("bharatpur",      1700, name="Bharatpur Load")
add_load("pokhara",        250, name="Pokhara Load")
add_load("kathmandu",       600, name="Kathmandu Load")
add_load("lapsifedi_220",       22, name="Lapsifedi Load")
add_load("new_hetauda_220",    1000, name="New Hetauda Load")
# if you used a 220 kV bus for Mirchaiya, use "new_mirchaiya_132"
add_load("new_mirchaiya_220", 110, name="New Mirchaiya Load")

add_load("inaruwa_220",        600, name="Inaruwa Load")
add_load("damak_132",            22, name="Damak Load")
add_load("bhaktapur",      112, name="Bhaktapur Load")
add_load("amarpur",         50, name="Amarpur Load")

add_load("chamelya",       150, name="Chamelya Load")
add_load("kaligandaki",    180, name="Kaligandaki Load")
add_load("kusma",          100, name="Kusma Load")
add_load("rahughat",       250, name="Rahughat Load")
add_load("new_damauli_220",    100, name="New Damauli Load")
add_load("m_marsyangdi",   155, name="M Marsyangdi Load")
add_load("barabise_220",        50, name="Barabise Load")
add_load("new_khimti_220",     260, name="New Khimti Load")
add_load("new_dhalkebar_220",  160, name="New Dhalkebar Load")

add_load("tingla",         100, name="Tingla Load")
add_load("dhungesaghu_132",    150, name="Dhungesaghu Load")
add_load("phidim",          75, name="Phidim Load")
add_load("godak",          100, name="Godak Load")

#interconnector Load
add_load("gorakhpur",          1000, name="Gorakhpur Load")
add_load("muzaffarpur",          1000, name="Muzaffarpur Load")
add_load("purnia",          1000, name="Purnia Load")



net.load
```
